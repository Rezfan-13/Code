```python
import pandas as pd
import plotly.graph_objects as go
from plotly.offline import init_notebook_mode, iplot
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("netflix_titlees.csv")
```

# 1) Tipe Konten

```python
ig,axes=plt.subplots(figsize=(12,6))

plt.pie(df.type.value_counts(),explode=(0.05,0.05),colors=['#DE3163','#40E0D0'],autopct="%1.1f%%",labels=df.type.value_counts().index)
plt.title("Movies vs TV Shows")
plt.legend()
plt.show()
```

**Hasil visualisasi menunjukan tipe konten yang paling banyak diproduksi pada platform Netflix adalah bertipe Movie**

# 2) Top 10 Directors

```python
fig,axes=plt.subplots(figsize=(12,6))
sns.barplot(
x=df.loc[df.director!="Unknown","director"].value_counts().values[0:10],
y=df.loc[df.director!="Unknown","director"].value_counts().index[0:10],palette="Set2")
plt.title("Top 10 Directors" )
plt.ylabel("Director")
plt.xlabel("Movies or TV Shows")
plt.show()
```

**Hasil visualisasi menunjukan 10 direktor teratas**

# 3) Top 5 Countries for Movies and TV Shows

```python
fig,ax=plt.subplots(figsize=(12,6))
sns.countplot(y=df.country,order=df.country.value_counts().index[0:5],palette="Set3")

plt.title("Top 5 Countries for Movies and TV Shows")
plt.xlabel("Number of Titles")
plt.ylabel("Country")
plt.show()
```

**Hasil visualisasi menunjukan 5 negara dengan penghasil film bertipe Movies dan Tv shows**

# 4) Content Rating of  Tv shows

```python
plt.figure(figsize=(12,6))
sns.countplot(x="rating",data=df,order= df['rating'].value_counts().index[0:14]);
plt.title('Content rating of shows', fontsize=18)
plt.xlabel('Ratings', fontsize=14)
plt.ylabel('Amount of content', fontsize=14);
```

**Hasil Visualisasi menunjukan rating konten berdasarkan tv shows**

# 5) Top 5 Movies Genre

```python
movie_genre = df.loc[df['type'] == 'Movie']
plt.figure(figsize=(12,6))
ax = sns.countplot(y='listed_in',data = movie_genre,order = movie_genre["listed_in"].value_counts().index[0:5],palette="Set1")
plt.title('Top 5 Movie Genre',fontsize = 15)
plt.xlabel('Total Count',fontsize = 15)
plt.ylabel('Genre',fontsize = 15)
```

**Hasil Visualisasi memperlihatkan 5 genre movies dengan total judul terbanyak**

# 6) Top 5 TV-Shows Genre

```python
tv_genre = df.loc[df['type'] == 'TV Show']
plt.figure(figsize=(12,6))
ax = sns.countplot(y='listed_in',data = tv_genre,order = tv_genre["listed_in"].value_counts().index[0:5],palette="Set2")
plt.title('Top 5 TV Shows Genre',fontsize = 15)
plt.xlabel('Total Count',fontsize = 15)
plt.ylabel('Genre',fontsize = 15)
```

**Hasil Visualisasi memperlihatkan 5 genre Tv-shows dengan total judul terbanyak**

# 7) Movies released by rating since 2016

```python
movie = df.loc[df['type']=='Movie']
movie_trends = movie[movie['release_year']>=2016]
x = movie_trends.groupby(['release_year','rating']).size().reset_index(name = 'Total')
df1 = x[x['Total'] >= 10]

plt.figure(figsize = (10,8))
sns.set_style('darkgrid')
sns.barplot(data = df1,x = 'release_year', y = 'Total',hue = 'rating',palette = 'Set2')
plt.title('Movies released by rating since 2016 ')
plt.xlabel('Year')
plt.ylabel('Total Count')
```

**Hasil Visualisasi menunjukan perilisan Movie  bedarsakan rating semenjak tahun 2016**

# 8) Tv Show released by rating since 2010

```python
movie = df.loc[df['type']=='TV Show']
movie_trends = movie[movie['release_year']>=2010]
x = movie_trends.groupby(['release_year','rating']).size().reset_index(name = 'Total')
df1 = x[x['Total'] >= 10]

plt.figure(figsize = (10,8))
sns.set_style('darkgrid')
sns.barplot(data = df1,x = 'release_year', y = 'Total',hue = 'rating',palette = 'Set2')
plt.title('Movies released by rating since 2016 and beyond(more than 10 releases per year)')
plt.xlabel('Year')
plt.ylabel('Total Count')
```

**Hasil Visualisasi menunjukan perilisan Tv show  bedasarkan rating semenjak tahun 2010**

# 9) Total Tv Show Season wise

```python
plt.figure(figsize= (10,5))
tv_show = df[df['type']== 'TV Show']

sns.countplot(x = 'duration',data = tv_show,palette = 'Blues_r', order = tv_show['duration'].value_counts().index)
plt.xticks(rotation = 90)
plt.xlabel("Seasons",fontsize = 15)
plt.ylabel("Total count",fontsize = 15)
plt.title("Total Tv Show Season",fontsize = 15)
```

**Hasil Visualisasi menunjukan total Tv show Season**

# 10) Movie Duration

```python
movie_duration = df.loc[df['type']=='Movie']

plt.figure(figsize=(12,10))

ax = sns.histplot(data = movie_duration,x = 'duration',bins = 50,kde = True,color = 'blue')
plt.title('Movie Duration',fontsize = 15)
plt.xlabel('Total Duration(In Mins)',fontsize = 15)
plt.ylabel('Total Movie Count',fontsize = 15)
```

**Hasil Visualisasi menunjukan durasi waktu film bertipe Movie**
