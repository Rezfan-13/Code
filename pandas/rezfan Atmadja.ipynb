Memuat dan mencetak informasi data set dasar:

### Menulis program Python untuk mendapatkan jumlah kasus yang dikonfirmasi, dirawat di rumah sakit, dipindahkan ke ICU, meninggal dan sembuh.

```python
a=df.groupby("Fecha")
b=a.sum()
c=b.sort_values(by=["Casos"],ascending=False)
c.head(1)
```

### Temukan jumlah terakhir dari kasus yang dikonfirmasi, meninggal dan dipulihkan oleh Komunitas:

```python
gk=df.groupby("CCAA")
rez=gk.sum()
tot=rez.sort_values(by=["Casos"],ascending=False)
tot.loc[:,["Casos","Fallecidos","Recuperados"]]
```

### Temukan jumlah kasus terbaru yang dikonfirmasi, dirawat di rumah sakit, dipindahkan ke ICU, meninggal dan pulih dari COVID-19 di Navarra

```python
a=df.sort_values(["Fecha"],ascending=False)
a[a["CCAA"]=="Navarra"].head(1)
```

### Temukan kolom baru dengan kasus aktif:

```python
import numpy as np
rez=df.replace(np.nan, 0)
rez["Aktif"]=rez["Casos"]+rez["Recuperados"]-rez["Fallecidos"]
rez.head()
```

### Temukan kapan kasus COVID-19 pertama dan kematian terjadi di Spanyol

```python
rez[rez["Casos"]>=1].head(1)
```

```python
b=rez.sort_values(["Fecha"],ascending=True)
b[b["Fallecidos"]==1].head(1)
```

### Menemukan kapan kematian pertama kali terjadi di setiap Komunitas

```python
a=rez[rez["Fallecidos"]>=1]

fan=a.groupby("CCAA")
at=fan.first()
at.sort_values("Fecha")
```
