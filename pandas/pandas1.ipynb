### Step 4. Select only the Goal column.

````
euro12.loc[0:,["Goals"]]
````

````
len(euro12["Team"])
````

### Step 6. View only the columns Team, Yellow Cards and Red Cards and assign them to a dataframe called discipline

````
disiplin=euro12.loc[0: , ["Team","Yellow Cards","Red Cards"]]
disiplin
````

### Step 7. Sort the teams by Red Cards, then to Yellow Cards

````
disiplin.sort_values(by=["Red Cards","Yellow Cards"],ascending=False)
````

````
disiplin["Yellow Cards"].mean()
````
