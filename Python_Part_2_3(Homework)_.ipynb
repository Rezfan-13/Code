```
rez=int(input(" enter no. of rows :"))
for i in range(1, rez+1):
  for j in range(1,i+1):
    if j==1:
      print(i,end=" ")
    else:
      x=i+(rez-j+1)
      print(x,end=" ")
      i=x
  print()
   
```
